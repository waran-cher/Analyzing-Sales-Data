```python
#Clean column name
    #Since the columns headers contain space and dash that can be difficult to type in later.  
    #It is better to change the column headers to be all lower case without space and dash.

#1.Create list of columns in dataframe
df_col = list(df.columns)

#2.Make all character lower case and remove space and dash using list comprehension
clean_col = [title.lower().replace(" ", "").replace("-", "") for title in df_col]

#3.Apply #2 list as new headers
df.columns = clean_col

df.head()
```

```python
# TODO 03 - your friend ask for `California` data, filter it and export csv for him

#Filter rows with condition that column 'state' exact contains California
ca = df[df['state'] == 'California']

#Use to_csv() function to write .csv file name sample_store_california.csv
ca.to_csv('sample_store_california.csv')

ca.head()
```

```python
# TODO 04 - your friend ask for all order data in `California` and `Texas` in 2017 (look at Order Date), send him csv file

#Filter row with..
    #1st condition: state is California or state is Taxas -> using | symbol as OR operator
    #2nd condition: orders that took place in year 2017
        #use .dt.year to filter only year 
#combine 1st condition and 2nd condition with & symbol
ca_tx_2017 = df[((df['state'] == 'California') | (df['state'] == 'Texas')) & (df['orderdate'].dt.year == 2017)] 

#Use to_csv() function to write .csv file name sample_store_california-texas_2017.csv
ca_tx_2017.to_csv('sample_store_california-texas_2017.csv')

ca_tx_2017.head()
```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017

#1.Create new dataframe that filter year only in 2017 name df2017
df2017 = df[df['orderdate'].dt.year == 2017]

#2.Use .agg() function to apply more than one aggregate function at once
df2017['sales'].agg(['sum', 'mean', 'std'])
```

```python
 # TODO 06 - which Segment has the highest profit in 2018

#1.Create new dataframe that filter year only in 2018 name df2018
df2018 = df[df['orderdate'].dt.year == 2018]

#2.Sum profit column in df2018 and group by segment column
#3.Reset index and sort column by descending order -> from highest to lowest profit
#4.Use head(1) function to get only the first row
df2018.groupby('segment')['profit'].sum().reset_index()\
    .sort_values('profit', ascending=False).head(1)
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019

#1.Create new dataframe that filter orderdate between 15 April 2019 - 31 December 2019 name sales1
sales1 = df[(df['orderdate'] >= '2019-4-15') & (df['orderdate'] <= '2019-12-31')]

#2.Sum sales column in sales1 and group by state column
#3.Reset index and sort sales column from lowest to highest (default:ascending order)
#4.Use head(5) function to show only first 5 rows
sales1.groupby('state')['sales'].sum().reset_index().sort_values('sales').head(5)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 

#1.Create new dataframe that filter year only in 2019 name df2019
df2019 = df[df['orderdate'].dt.year == 2019]

#2.Filter row in df2019 that region is in west or central using .query() function
    #then assign filtered dataframe to wc_2019
wc_2019 = df2019.query("region == 'West' | region == 'Central'")

#3.Get total sales of df2019 and wc_2019 -> df2019['sales'].sum() and wc_2019['sales'].sum()
total_sales_2019 = df2019['sales'].sum()
total_sales_wc_2019 = wc_2019['sales'].sum()

#4.Find proportion 
prop = (total_sales_wc_2019 / total_sales_2019)*100

#5.Print result with 2 digit decimal using .round(decimals=2)
print(prop.round(decimals=2),'%')

# q25 = sales2019['sales'].quantile(0.25)
# q50 = sales2019['sales'].quantile(0.5)
# q75 = sales2019['sales'].quantile(0.75)

# print(q25,q50, q75, sep='\n')
```

```python

# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020

#1.Create new dataframe that filtered in 2019-2020 name df2019_2020
df_2019_2020 = df[(df['orderdate'].dt.year == 2019) | (df['orderdate'].dt.year == 2020)]

#2.Find total order of all product in 2019-2020  : count only order not quantity
#   2.1 Count values productname column in df_2019_2020 
#   2.2 Sort values in descending order and reset index
#   2.3 Assign results to new variable name total_pdsales_2019_2020
total_pdorder_2019_2020 = df_2019_2020['productname'].value_counts().sort_values(ascending=False).reset_index()
#   2.4 Change columns name from index(old) to productname(new) and from productname(old) to order(new)
total_pdorder_2019_2020 = total_pdorder_2019_2020.rename(columns = {'index':'productname', 'productname':'order'})

#3.Find total sales of all product in 2019-2020 
#   3.1 Sum sales column in df_2019_2020 and group by productname column
#   3.2 Sort values in descending order and reset index
#   3.3 Assign results to new variable name total_pdsales_2019_2020
total_pdsales_2019_2020 = df_2019_2020.groupby('productname')['sales'].sum().sort_values(ascending=False).reset_index()

#4.Concat 2 dataframes use pd.concat() function : axis=1 is to concatenating along the columns
#   4.1 .head(10) to show only the first 10 rows
top10_product = pd.concat([total_pdorder_2019_2020, total_pdsales_2019_2020],axis=1).head(10)

top10_product
```

```python

#Profit over the years

#1.Sum profit group by year in orderdate column
total_profit_years = df.groupby(df['orderdate'].dt.year)['profit'].sum().reset_index()

#2.Change column name from orderdate(old) to year(new)
total_profit_years = total_profit_years.rename(columns = {'orderdate':'year'})

#3.convert datatype in year column from int64 to string
total_profit_years['year'] = total_profit_years['year'].astype('str')
    #If don't convert to string the year axis in plot will have a decimal

#4. plot line graph : x axis = year column and y axis = profit column
total_profit_years.plot.line(x='year', y='profit')
```
